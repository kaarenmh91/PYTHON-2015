```python
%matplotlib inline
import matplotlib.pyplot as plt
import sys, os, re, time
import numpy as np
import cv2
```

```python
import android
 
```

```python
pic = os.path.join('.','ZuzanaB.jpg')#ZuzanaB.jpg')
```

```python
img=cv2.imread(pic)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
caras_cascada=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
caras = caras_cascada.detectMultiScale(gray, 1.2, 2)
```

```python
print "Encontramos {0} faces!".format(len(caras))
```

```python
for (x,y,w,h) in caras:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
```
